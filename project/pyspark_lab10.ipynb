```python
# --- 1. Импорт и создание сессии Spark ---
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Lab9_PySpark_DataFrame") \
    .getOrCreate()

# --- 2. Загрузка CSV-файла ---
path = "vgsales.csv"

df = spark.read.csv("vgsales.csv", header=True, inferSchema=True).limit(500)

# --- 3. Первые операции с DataFrame ---
print("=== Первые строки датасета ===")
df.show(5)

# Проверка структуры таблицы
print("=== Схема данных ===")
df.printSchema()

# Выбор отдельных столбцов
print("=== Названия игр и платформ ===")
df.select("Name", "Platform").show(5)

# Фильтрация данных по условию
print("=== Игры с продажами более 5 млн ===")
df.filter(df["Global_Sales"] > 5).show(5)

# Сортировка по убыванию продаж
print("=== Топ по мировым продажам ===")
df.orderBy(df["Global_Sales"].desc()).show(5)

# Базовая статистика
print("=== Статистические показатели продаж ===")
df.describe(["Global_Sales", "NA_Sales", "EU_Sales"]).show()
```
